```python
import pandas as pd
import glob
import os
```

```python
path = r'mtree-csv'
all_files = glob.glob(path + "/*.csv")

df_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df_list.append(df)

combined_df = pd.concat(df_list, axis=0, ignore_index=True)

combined_df.to_csv("mtree-csv/combined_mtree.csv", index=False)
```

```python
progenitor_info = pd.read_csv('mtree-csv/combined_mtree.csv')

snapshot_files = glob.glob('halo-csv/GadgetX-NewMDCLUSTER_0001.snap_*.csv')

output_dir = 'combined-csv/'
```

```python
# Process each snapshot file
for snapshot_file in snapshot_files:

    snapshot = pd.read_csv(snapshot_file)

    merged = pd.merge(snapshot, progenitor_info, on='ID', how='left')

    output_file = os.path.join(output_dir, os.path.basename(snapshot_file).replace('.csv', '_combined.csv'))

    merged.to_csv(output_file, index=False)
```

```python
# Loop over all files in the output directory
for filename in os.listdir(output_dir):

    # Only process .csv files
    if filename.endswith('.csv'):

        # Get the path to the file
        file_path = os.path.join(output_dir, filename)

        # Read the csv file into a pandas dataframe
        data = pd.read_csv(file_path)

        # Fill missing values
        data['numProgenitors'].fillna(0, inplace=True)
        data['ProgenitorsID'].fillna('-1', inplace=True)

        # Convert numProgenitors to integer
        data['numProgenitors'] = data['numProgenitors'].astype(int)
        
        # Convert ProgenitorsID to a list of integers
        data['ProgenitorsID'] = data.apply(lambda row: -1 if row['numProgenitors'] == 0 or row['ProgenitorsID'] == -1.0 
                                              else [int(row['ProgenitorsID'])] if row['numProgenitors'] == 1 
                                              else [int(i) for i in str(row['ProgenitorsID']).split(",") if i.isdigit()], axis=1)
        
        # Extract redshift from filename and add as a new column
        redshift = float(filename.split('z')[1].split('_')[0])
        data['redshift'] = redshift

        # Extract snapshot number from filename and add as a new column
        snapshot_number = int(filename.split('snap_')[1].split('.')[0])
        data['snapshot'] = snapshot_number

        # Save the data back to the same file (overwrite)
        data.to_csv(file_path, index=False)
```
