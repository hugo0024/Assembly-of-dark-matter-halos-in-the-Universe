```python
import pandas as pd
import os
```

```python
file_names = os.listdir('../Preprocess/GadgetX-NewMDCLUSTER-1/halo-csv/')
```

```python
dataframes = []
for file_name in file_names:
    df = pd.read_csv(f'../Preprocess/GadgetX-NewMDCLUSTER-1/halo-csv/{file_name}')
    dataframes.append(df)
```

```python
sample_data = pd.concat(dataframes)
```

```python
import matplotlib.pyplot as plt

# Set up the figure and axes
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plot histograms
sample_data['Mvir'].plot(kind='hist', bins=50, ax=axes[0], color='skyblue', edgecolor='black')
sample_data['npart'].plot(kind='hist', bins=50, ax=axes[1], color='salmon', edgecolor='black')

# Set titles and labels
axes[0].set_title('Distribution of Virial Mass (Mvir)')
axes[0].set_xlabel('Mvir')
axes[0].set_ylabel('Frequency')

axes[1].set_title('Distribution of Number of Particles (npart)')
axes[1].set_xlabel('npart')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker
```

```python
plt.figure(figsize=(10, 8))
sc = plt.scatter(sample_data['Xc'], sample_data['Yc'], c=sample_data['Mvir'], s=50, cmap='viridis', edgecolor='k', alpha=0.7)

# Creating the colorbar and adjusting the font size
cbar = plt.colorbar(sc, label='Virial Mass ($M_{vir}$)')
cbar.ax.set_ylabel('Virial Mass ($M_{vir}$)', fontsize=20)
cbar.ax.tick_params(labelsize=20)

# Setting the title, x and y labels with specified font sizes and weights
plt.title('Halos Distribution at snapshot 30', fontweight='bold', fontsize=20)
plt.xlabel('$X_c$', fontsize=25)
plt.ylabel('$Y_c$', fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Using scientific notation for x and y axis
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
ax.yaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
ax.ticklabel_format(style='sci', scilimits=(0,0), axis='both')

# Other formatting details
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Calculate velocity magnitudes for all halos
sample_data['velocity_magnitude'] = (sample_data['VXc']**2 + sample_data['VYc']**2 + sample_data['VZc']**2)**0.5

plt.figure(figsize=(12, 8))
sns.histplot(sample_data['velocity_magnitude'], bins=50, color='mediumpurple', kde=True)

# Setting the title, x and y labels with specified font sizes and weights
plt.title('Distribution of Velocity Magnitudes', fontweight='bold', fontsize=25)
plt.xlabel('Velocity Magnitude', fontsize=25)
plt.ylabel('Frequency', fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
